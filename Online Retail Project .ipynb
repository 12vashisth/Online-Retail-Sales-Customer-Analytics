```python
#  Convert InvoiceDate to datetime format (if not already)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

#  Create TotalAmount (Revenue per transaction)
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
```

```python
#  Extract Time-Based Features
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
df['Hour'] = df['InvoiceDate'].dt.hour
df['Weekday'] = df['InvoiceDate'].dt.day_name()

# Create Date Column (useful for daily analysis)
df['Date'] = df['InvoiceDate'].dt.date
```

```python
#  Create Customer-Level Features
customer_df = df.groupby('CustomerID').agg({
    'TotalAmount': 'sum',          # Total money spent
    'InvoiceNo': 'nunique'         # Number of unique transactions
}).reset_index()

customer_df.rename(columns={
    'TotalAmount': 'Total_Spending',
    'InvoiceNo': 'Total_Transactions'
}, inplace=True)
```

```python
# RFM Feature Engineering
# 1. Ensure InvoiceDate is datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# 2. Create a reference date (latest date in dataset + 1 day)
reference_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)
```

```python
# 3. Create RFM features
rfm_df = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,  # Recency
    'InvoiceNo': 'nunique',                                   # Frequency
    'TotalAmount': 'sum'                                      # Monetary
}).reset_index()
```

```python
# 4. Rename columns
rfm_df.rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalAmount': 'Monetary'
}, inplace=True)

```

```python
# 5. Display top rows
rfm_df.head()
```

```python
# Create RFM Scores using quartiles
rfm_df['R_Score'] = pd.qcut(rfm_df['Recency'], 4, labels=[4,3,2,1])
rfm_df['F_Score'] = pd.qcut(rfm_df['Frequency'].rank(method='first'), 4, labels=[1,2,3,4])
rfm_df['M_Score'] = pd.qcut(rfm_df['Monetary'], 4, labels=[1,2,3,4])

# Combine scores into one RFM score
rfm_df['RFM_Score'] = (
    rfm_df['R_Score'].astype(str) +
    rfm_df['F_Score'].astype(str) +
    rfm_df['M_Score'].astype(str)
)

# View top customers
rfm_df.sort_values(by='RFM_Score', ascending=False).head()
```

```python
# ================================
# Customer Segmentation using RFM
# ================================

def segment_customer(row):
    r, f, m = int(row['R_Score']), int(row['F_Score']), int(row['M_Score'])
    
    if r >= 3 and f >= 3 and m >= 3:
        return 'Best Customers'
    elif f >= 3 and m >= 3:
        return 'Loyal Customers'
    elif r >= 3:
        return 'Recent Customers'
    elif f >= 3:
        return 'Frequent Customers'
    else:
        return 'Low Value Customers'

rfm_df['Customer_Segment'] = rfm_df.apply(segment_customer, axis=1)
```
